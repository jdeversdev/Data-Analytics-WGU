```python
# Importing necessary libraries for data cleaning and others

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Load the dataset
data_csv = '/content/churn_raw_data.csv'
data = pd.read_csv(data_csv)

# Data Assessment
# Display the first few rows of the dataset to get an overview
data.head()
```

```python
# Profiling the data

# Checking the shape of the dataset
shape = data.shape

# Checking data types of each column
data_types = data.dtypes

# Checking for missing values
missing_values = data.isnull().sum()

# Checking for duplicates
duplicate_count = data.duplicated().sum()

shape, data_types, missing_values, duplicate_count
```

```python
# Data Mitigation

# Treating Missing Values

# For simplicity, we'll use median imputation for continuous variables and mode imputation for categorical variables

# List of columns with missing values
cols_with_missing = missing_values[missing_values > 0].index.tolist()

# Continuous columns with missing values
continuous_cols_missing = ['Children', 'Age', 'Income', 'Tenure', 'Bandwidth_GB_Year']

# Categorical columns with missing values
categorical_cols_missing = list(set(cols_with_missing) - set(continuous_cols_missing))

# Median imputation for continuous columns
for col in continuous_cols_missing:
    median_value = data[col].median()
    data[col].fillna(median_value, inplace=True)

# Mode imputation for categorical columns
for col in categorical_cols_missing:
    mode_value = data[col].mode()[0]
    data[col].fillna(mode_value, inplace=True)

# Verify if missing values are treated
data[cols_with_missing].isnull().sum()
```

```python
#Fixing data types
# Convert selected columns to appropriate data types
data['Income'] = data['Income'].astype(int)
data['Children'] = data['Children'].astype(int)
data['Tenure'] = data['Tenure'].astype(int)
data['Bandwidth_GB_Year'] = data['Bandwidth_GB_Year'].astype(int)
data['Age'] = data['Age'].astype(int)

# Check and print data types
data_types = data.dtypes
print("Data Types After Conversion:")
print(data_types)
```

```python
# Detecting Outliers using Boxplots
import matplotlib.pyplot as plt
import seaborn as sns

# Setting up the figure and axes
plt.figure(figsize=(15,10))

# Plotting boxplots for continuous columns to detect outliers
for i, col in enumerate(continuous_cols_missing, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(y=data[col])
    plt.title(f'Boxplot of {col}')
    plt.ylabel(col)

plt.tight_layout()
plt.show()
```

```python
# Using the Percentile Method to detect and treat outliers

# Define the upper and lower bounds using percentiles
lower_bound = data.quantile(0.01)
upper_bound = data.quantile(0.99)

# Replace outliers with NaN
data_outliers_treated = data.copy()
for column in continuous_cols_missing:
    data_outliers_treated[column] = np.where(
        (data_outliers_treated[column] < lower_bound[column]) |
        (data_outliers_treated[column] > upper_bound[column]),
        np.nan,
        data_outliers_treated[column]
    )

# Impute NaN values with median
for column in continuous_cols_missing:
    median_value = data_outliers_treated[column].median()
    data_outliers_treated[column].fillna(median_value, inplace=True)

# Displaying boxplots before and after treating outliers
import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(15, 10))

# Before treating outliers
data[continuous_cols_missing].boxplot(ax=axes[0])
axes[0].set_title('Boxplots Before Treating Outliers')
axes[0].set_xticklabels(axes[0].get_xticklabels(), rotation=45, ha='right')

# After treating outliers
data_outliers_treated[continuous_cols_missing].boxplot(ax=axes[1])
axes[1].set_title('Boxplots After Treating Outliers')
axes[1].set_xticklabels(axes[1].get_xticklabels(), rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

# Select the numerical columns to normalize
numerical_cols = ['Income']

# Initialize the StandardScaler
scaler = StandardScaler()

# Fit and transform the selected columns
data_filtered_ny[numerical_cols] = scaler.fit_transform(data_filtered_ny[numerical_cols])

# Print the resulting DataFrame
print(data_filtered_ny)
```

```python
# Perform one-hot encoding for the 'InternetService' column
data_encoded = pd.get_dummies(data_filtered_ny, columns=['InternetService'], drop_first=True)

print(data_encoded)
```

```python
# Normalize the data
scaler = StandardScaler()
data_normalized = pd.DataFrame(scaler.fit_transform(data_encoded), columns=data_encoded.columns)

print(data_normalized)
```

```python
from tabulate import tabulate

# Perform PCA
pca = PCA()
data_pca = pca.fit_transform(data_normalized)

# Access PCA results, including loadings
loadings = pca.components_

# Identify the corresponding column names and their respective loadings values
column_names = data_encoded.columns

# Create a loading matrix
loading_matrix = {}

# Populate the loading matrix
for i, name in enumerate(column_names):
    loading_values = loadings[:, i]
    loading_matrix[name] = loading_values

# Organize the loading matrix for tabulate
table_data = []
for name in column_names:
    row = [name] + [f"{value:.6f}" for value in loading_matrix[name]]  # Change to '.6f' for at least three digits after the decimal point
    table_data.append(row)

# Print the organized loading matrix using tabulate
headers = [''] + [f"PC{j + 1}" for j in range(len(loadings))]
print(tabulate(table_data, headers, tablefmt="fancy_grid"))
```

```python
import matplotlib.pyplot as plt

# Perform PCA
pca = PCA()
data_pca = pca.fit_transform(data_normalized)

# Access PCA results, including explained variance
explained_variance = pca.explained_variance_ratio_

# Create a scree plot
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(explained_variance) + 1), explained_variance, marker='o', linestyle='--')
plt.title("Scree Plot")
plt.xlabel("Number of Principal Components")
plt.ylabel("Explained Variance")
plt.grid()

# Add a marker for the selected number of principal components
# 5 principal components
plt.axvline(x=5, color='red', linestyle='--', label="Selected Components (e.g., 2)")

# Show the plot
plt.legend()
plt.show()
```
