```python
# Import necessary libraries
import pandas as pd

# Load the dataset

data_csv = '/content/medical_clean.csv'
data = pd.read_csv(data_csv)

# Display the first few rows of the dataset
data.head()

# Checking the shape of the dataset
data_shape = data.shape
print("Data shape:", data_shape)

# Checking data types of each column
data_types = data.dtypes
print("Data types:", data_types)

# Checking for missing values
missing_values = data.isnull().sum()
print("Missing values:", missing_values)

# Checking for duplicates
duplicate_count = data.duplicated().sum()
print("Duplicate count:", duplicate_count)
```

```python
import pandas as pd
from scipy.stats import ttest_ind

# Necessary columns
selected_columns = ['Age', 'ReAdmis']
df = data[selected_columns]

# Separate data into two groups based on ReAdmis
readmitted = df[df['ReAdmis'] == 'Yes']['Age']
not_readmitted = df[df['ReAdmis'] == 'No']['Age']

# Perform t-test
t_stat, p_value = ttest_ind(readmitted, not_readmitted, equal_var=False)

# Display results
print("T-Test Results:")
print(f'T-statistic: {t_stat}')
print(f'P-value: {p_value}')

# Interpretation
if p_value < 0.05:
    print("The difference in average age between readmitted and not readmitted patients is statistically significant.")
else:
    print("There is no significant difference in average age between readmitted and not readmitted patients.")
```

```python
# D. Bivariate Statistics

# Bivariate Statistics for Continuous-Continuous Relationship
correlation_matrix = data[['Age', 'VitD_levels']].corr()
print("Correlation Matrix for Age and VitD_levels:")
print(correlation_matrix)

# Bivariate Statistics for Categorical-Continuous Relationship
average_age_by_gender = data.groupby('Gender')['Age'].mean()
print("\nAverage Age by Gender:")
print(average_age_by_gender)

average_vitD_by_highblood = data.groupby('HighBlood')['VitD_levels'].mean()
print("\nAverage VitD_levels by HighBlood:")
print(average_vitD_by_highblood)

# Bivariate Statistics for Categorical-Categorical Relationship
cross_tab_gender_highblood = pd.crosstab(data['Gender'], data['HighBlood'])
print("\nCross-tabulation between Gender and HighBlood:")
print(cross_tab_gender_highblood)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# D. Bivariate Statistics

# Set the style for the plots
sns.set(style="whitegrid")

# Scatter plot for Age and VitD_levels
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='VitD_levels', data=data, color='skyblue')
plt.title('Scatter Plot: Age vs. VitD_levels')
plt.xlabel('Age')
plt.ylabel('VitD_levels')
plt.show()
```

```python
# Box plot for Average VitD_levels by HighBlood
plt.figure(figsize=(8, 5))
sns.boxplot(x='HighBlood', y='VitD_levels', data=data, palette='muted')
plt.title('Box Plot: HighBlood vs. VitD_levels')
plt.xlabel('HighBlood')
plt.ylabel('VitD_levels')
plt.show()
```

```python
# Stacked bar chart for Cross-tabulation between Gender and HighBlood
plt.figure(figsize=(10, 6))
cross_tab_plot = sns.countplot(x='Gender', hue='HighBlood', data=data, palette='pastel')
cross_tab_plot.legend(title='HighBlood')
plt.title('Cross-tabulation: Gender vs. HighBlood')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```
